```python
from abc import ABC, abstractmethod

class BankAccount(ABC):              #abstract class
  @abstractmethod
  def transaction(self):           #abstract method
      pass

#Encapsulation
# creating a class name Account that is used to store account details using public, protected, and private variables
class Account(BankAccount):
    def __init__(self, account_holder_name, account_number, account_balance, pin):
      self.account_holder_name = account_holder_name      #public
      self._account_number = account_number               #protected
      self.__account_balance = account_balance            #private
      self.__pin = pin                                    #private
      self.transaction_count = 0                          #public
      self.transaction_history = []                       #public

    def show_name(self):
      print("Account Holder Name:", self.account_holder_name)

    def _show_account_number(self):
      print("Account Number:", self._account_number)

    def __show_balance(self):
      print("Account Balance:", self.__account_balance)

    def display(self):                                   #creating display method to access the private variable(__)
      self.__show_balance()

    def transaction(self):
      print("Basic bank transaction")

    def verify_pin(self, entered_pin):
      return self.__pin == entered_pin

#Inheritance
#Creating the ATM class that inherits the Account
class ATM(Account):
    bank_name = "Bank Of India"
    MAX_WITHDRAW_LIMIT = 5000                    #maximum withdrawal limit

    @classmethod
    def show_bank_name(cls):
      print("WELCOME TO", cls.bank_name, "ATM")

    #used to show the account holder name and account number
    #method overriding
    def transaction(self):                           #implementing the abstract method
      print("--------------------------------")
      ATM.show_bank_name()
      print("--------------------------------")
      print("Account Holder:", self.account_holder_name)
      print("Account Number:", self._account_number)

    #static method
    @staticmethod
    def validate_amount(amount):
      return amount > 0

    #used to check the balance
    def check_balance(self):
      print("Current Balance:", self._Account__account_balance)
      self.transaction_count += 1
      self.transaction_history.append("Checked Balance")

    #used to deposit money in the account
    def deposit(self, amount):
      if ATM.validate_amount(amount):
        self._Account__account_balance += amount
        print("Deposit Successful")
        print("Updated Balance:", self._Account__account_balance)
        self.transaction_count += 1
        self.transaction_history.append(f"Deposited {amount}")
      else:
        print("Invalid deposit amount")

    #it check the given withdraw amount is in  given limit or not
    def withdraw(self, amount):
      if amount > ATM.MAX_WITHDRAW_LIMIT:
          print("Withdrawal limit exceeded (Max:", ATM.MAX_WITHDRAW_LIMIT, ")")
          return
      #used to withdraw the amount
      if ATM.validate_amount(amount):
          if amount <= self._Account__account_balance:
            self._Account__account_balance -= amount
            print("Withdrawal Successful")
            print("Updated Balance:", self._Account__account_balance)
            self.transaction_count += 1
            self.transaction_history.append(f"Withdrawn {amount}")
          else:
            print("Insufficient balance")
      else:
        print("Invalid withdrawal amount")
    #used to show the account details
    def account_details(self):
      self.show_name()
      self._show_account_number()
      self.display()

    #used to show the total transaction
    def show_transaction_history(self):
      print("\nTransaction History:")
      for t in self.transaction_history:
          print("-", t)
      print("Total Transactions:", self.transaction_count)



atm = ATM("Naveen", 123456789, 10000, 1234)

access_granted = True

#pin verification system
pin = int(input("Enter your ATM PIN: "))
if not atm.verify_pin(pin):
  print("Incorrect PIN. Access Denied.")
  access_granted = False


if access_granted:
  atm.transaction()

    #Use while True and repeat menu until Exit
  while True:
      print("\nATM MENU")
      print("1. Check Balance")
      print("2. Deposit Money")
      print("3. Withdraw Money")
      print("4. Account Details")
      print("5. Transaction History")
      print("6. Exit")

      choice = input("Enter your choice: ")

      if choice == "1":                            #check balance
        atm.check_balance()

      elif choice == "2":                          #deposit money
        amount = float(input("Enter deposit amount: "))
        atm.deposit(amount)

      elif choice == "3":                          #withdraw money
        amount = float(input("Enter withdrawal amount: "))
        atm.withdraw(amount)

      elif choice == "4":                           #account details
        atm.account_details()

      elif choice == "5":
        atm.show_transaction_history()         #transaction count
                                                   #transaction history

      elif choice == "6":                               #exit the loop
        print("Thank you for using", ATM.bank_name, "ATM.")
        print("Visit Again")
        break

      else:
        print("Invalid choice. Please try again.")
```
